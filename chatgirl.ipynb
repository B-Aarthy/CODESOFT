```python


def chatgirl():
    print("Hello! I am your friendly chatgirl. Type 'exit' to end the conversation.")

    while True:
        user_input = input("You: ").lower()
        if user_input == "exit":
            print("Chatgirlt: Goodbye! Have a nice day.")
            break


        elif "hello" in user_input or "hi" in user_input:
            print("Chatgirl: Hello! How can I help you today?")


        elif "your name" in user_input:
            print("Chatgirl: I am a simple rule-based chatgirl.")


        elif "how are you" in user_input:
            print("Chatgirl:  I'm doing fine! How about you?. Thanks for asking dear")


        elif "time" in user_input:
            from datetime import datetime
            now = datetime.now()
            print(f"Chatgirl: Current time is {now.strftime('%H:%M:%S')}")

        elif "date" in user_input:
            from datetime import datetime
            today = datetime.today()
            print(f"Chatgirl: Today's date is {today.strftime('%Y-%m-%d')}"

        else:
            print("Chatgirl: Sorry, I didn't understand that. Can you rephrase?")

chatgirl()
```
